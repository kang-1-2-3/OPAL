```python
import os 
import pykitti 
import numpy as np

pc_dir = '/home/kang/dataset/data_odometry_velodyne/dataset'
label_dir = '/home/kang/dataset/dataset'

seq_list = ['00', '01', '02', '03', '04', '05', '06', '07', '08', '09', '10']
for seq in seq_list:
    dataset = pykitti.odometry(pc_dir, sequence=seq)
    for pc_file_path in dataset.velo_files:
        pc = np.fromfile(pc_file_path, dtype=np.float32).reshape(-1, 4)[:,:3]
        # print(os.path.basename(pc_file_path))
        label_file_path = os.path.join(label_dir, 'sequences', seq, 'labels', os.path.splitext(os.path.basename(pc_file_path))[0] + '.label')
        label = np.fromfile(label_file_path, dtype=np.uint32)
        label = label.reshape((-1))
        # extract building points
        building_points = pc[label == 50]
        print(building_points.shape)
        break
```
